```python
The subplot() function in Matplotlib is used to create a grid of subplots within 
a single figure. Each subplot is essentially a separate plot or chart that allows 
you to display different sets of data or related visualizations within the same 
figure window. This function is especially useful when you want to create complex 
and customized layouts for your plots, such as arranging multiple plots in rows and columns
```

```python
import matplotlib.pyplot as plt
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1)
plt.plot(x1, y1, label='Line 1', color='blue')
plt.title('Line Plot 1')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x2, y2, label='Line 2', color='green')
plt.title('Line Plot 2')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x3, y3, label='Line 3', color='red')
plt.title('Line Plot 3')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(x4, y4, label='Line 4', color='purple')
plt.title('Line Plot 4')
plt.legend()

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()

```

```python
A bar plot, also known as a bar chart or bar graph, is a type of data visualization 
that displays categorical data with rectangular bars. Each bar represents a category, 
and the length or height of the bar corresponds to the value or frequency of that category.

It is used:
1.Comparing Categories
2.Visualizing Frequencies
3.Tracking Changes Over Time
4.Showing Relationships
5.Comparing Subcategories
```

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
```

```python
plt.barh(company,profit)
```

```python
A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays key summary statistics, such as the median, quartiles (25th and 75th percentiles), and potential outliers, using a box and whisker diagram. 

Box plots are used for the following purposes:
1.Visualizing Data Distribution
2.Identifying Outliers
3.Displaying Data Spread
4.Highlighting Central Tendency
5.Comparing Distributions
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]
labels = ['Box 1', 'Box 2']
plt.boxplot(data, labels=labels)
```
